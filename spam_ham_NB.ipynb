##**Problem Statement**

The SMS Spam Collection is a set of SMS tagged messages that have been collected for SMS Spam research. It contains one set of SMS messages in English of 5,574 messages, tagged according to being ham (legitimate) or spam.

```python
import numpy as np
import pandas as pd
import os
```

##Structure The data frame

```python

df = pd.read_csv('spam.csv', encoding="ISO-8859-1")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
# rename columns

df.columns = ['labels', 'sms']
```

## Describe the sms structure

```python
df.groupby('labels').describe()
```

# b)Remove punctuations, special characters and stopwords from the text in ‘sms’ column. Convert the text to lower case.


```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# remove special character
df['sms'] = df['sms'].str.replace('[#,@,&]', '')
# remove punctuatuons
df['sms'] = df['sms'].str.replace('[^\w\s]','')
# remove stop words
stop_words = stopwords.words('english')
df['sms'] = df['sms'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
# lower case
df['sms'] = df['sms'].str.lower()
df.head(5)
```

# Create two objects x and y

```python
X = df['sms']
y = df['labels']
```

# Train and test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# Appy count vectorizer and MultinomialNB to predict

```python
#from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X_train_vector = vectorizer.fit_transform(X_train)
X_test_vector = vectorizer.transform(X_test)
model = MultinomialNB()
model.fit(X_train_vector,y_train)
y_predictions = model.predict(X_test_vector)
```

## Evaluation of model

```python
# valuation
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_predictions))
```
