```python
import numpy as np
import pandas as pd
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

# Let's Plan

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False, handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python

# Scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature selection
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Pipeline Vs make_pipeline

Pipeline requires naming of steps, make_pipeline does not.

(Same applies to ColumnTransformer vs make_column_transformer)

```python
# Alternate Syntax
pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```

# Explore the Pipeline

```python
# Code here
pipe.named_steps
```

```python
# prompt: get input columns type from trf1 simpleimputer

# Access the named steps of the pipeline
named_steps = pipe.named_steps

# Access the SimpleImputer transformer
imputer_transformer = named_steps['trf1']

# Access the transformers within the ColumnTransformer
transformers = imputer_transformer.transformers_

# Iterate through the transformers and print their column indices and input column types
for name, transformer, columns in transformers:
  print(f"Transformer Name: {name}")
  print(f"Column Indices: {columns}")

  # Access the underlying SimpleImputer
  if isinstance(transformer, SimpleImputer):
      # Get the input column types
      if columns:  # Check if any column indices are provided
          input_cols = X_train.iloc[:,columns].dtypes.to_dict()
          print(f"Input Column Types: {input_cols}")
      else:
          print("No column indices provided for this transformer")
```

```python
pipe.named_steps['trf1']
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross Validation using Pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

# GridSearch using Pipeline

```python
# gridsearchcv
params = {
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```
