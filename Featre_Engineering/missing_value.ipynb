```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')
df.sample(5)
```

Find missing columns

```python
df.isnull().sum()
```

Age has so many missing values
Lets fix it now

## Imputation technique
   Mean median mode

```python
df['Age'].fillna(df['Age'].median())
df['Age'].fillna(df['Age'].mean())
df['Age'].fillna(df['Age'].mode())
```

### Some other impurity technique where mean median mode is not applicable

```python
df2 = pd.read_csv('sample_missing_dataset.csv')
df2.sample(5)
df2.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
# Daily sales is hughly skewed data so lets apply
# Separate missing and non-missing
train = df2[df2["Daily_Sales"].notna()]
test = df2[df2["Daily_Sales"].isna()]
features = ["Ad_Spend", "Website_Visits"]

# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train[features], train["Daily_Sales"])

# Predict missing values
predicted_dailysales = rf.predict(test[features])
type(predicted_dailysales)
print(predicted_dailysales)
# website visits
train = df2[df2["Website_Visits"].notna()]
test = df2[df2["Website_Visits"].isna()]
features = ["Ad_Spend", "Daily_Sales"]
rf.fit(train[features], train["Website_Visits"])
predicted_websitevisits = rf.predict(test[features])
print(predicted_websitevisits)
```
