```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

# Load dataset
wine = load_wine()
X = wine.data

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled.shape)
```

```python
# Apply SVD
U, S, VT = np.linalg.svd(X_scaled)

# Reduce dimensions (keep first 3 components)
X_reduced = np.dot(U[:, :3], np.diag(S[:3]))

print(X_reduced.shape)
```

Lets ee scatter plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

y = wine.target

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    X_reduced[:,0],
    X_reduced[:,1],
    X_reduced[:,2],
    c=y
)

ax.set_xlabel("SVD Component 1")
ax.set_ylabel("SVD Component 2")
ax.set_zlabel("SVD Component 3")

plt.title("3D Visualization of First 3 SVD Components")
plt.show()
```
