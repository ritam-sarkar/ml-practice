```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load dataset
wine = load_wine()
X = wine.data
y = wine.target
print(X.shape)
```

It has 178 samples and 13 features

```python
# Step 1: Feature Scaling (IMPORTANT for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 2: Apply PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
print(X_pca.shape)
```

It means 13 features come down to only 3 features

```python
# Step 3: Check explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    X_pca[:,0],
    X_pca[:,1],
    X_pca[:,2],
    c=y
)

ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")

plt.title("3D PCA Visualization")
plt.show()
```

If you see  Explained Variance Ratio: [0.36198848 0.1920749 0.11123631]
It means total variance captured 0.3619 + 0.1920 + 0.1112 = 0.665
so converting it to 3 features can capture upto 0.665 variance or 66.5 % , we can add more component but can never reach 1 untill
adding all 13 features


### Automatic PCA selection

```python
pca = PCA(n_components=0.95)
X_pca2 = pca.fit_transform(X_scaled)

print(pca.n_components_)
print(X_pca2.shape)
```

To reach 0.95 or 95% we need 10 components
Generally 60 to 70% is moderate so lets go with 3 components
